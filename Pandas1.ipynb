```python
import pandas as pd
import numpy as np
data = pd.DataFrame({"Product_Name":["Iphone 15 ","SamsungS24 ","Oppo F26 ", "Iphone 17 pro ", "Vivo v8 "],
                      "Price":[78000,67000,45000,110000,34000]})
print(data)
arr = data["Price"]
avg = np.mean(arr)
print(" average price :- ", avg)

def products_above_average(data, avg):
    print("\nProducts priced above the average:")
    
    for i in range(len(data)):
        if data.loc[i, "Price"] > avg:
            print(data.loc[i, "Product"],
                  "- Price:", data.loc[i, "Price"])

products_above_average(data, avg)
```
