```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("StudentsPerformance1.csv")
data
```

```python
data.isnull().sum()
```

```python
mean = np.mean(data["Exam_Score"])
print(mean)
median = np.median(data["Exam_Score"])
print(median)
std = np.std(data["Exam_Score"])
print(std)
```

```python
plt.hist(data["Exam_Score"])
plt.xlabel(" Marks ")
plt.title(" Histogram ")
plt.show()
```

```python
plt.scatter(data["Hours_Studied"],data["Exam_Score"])
plt.xlabel(" Hours ")
plt.ylabel(" score ")
plt.title(" Scatter Plot")
plt.show()
```

```python
x = data[["Hours_Studied"]]
y = data["Exam_Score"]
```

```python
Lr = LinearRegression()
Lr.fit(x,y)
pred_values = Lr.predict(x)
pred_values
```

```python
plt.scatter(x,y,)
plt.plot(x,pred_values,linewidth= 6,color='red')
```
