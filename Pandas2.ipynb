```python
import numpy as np
import pandas as pd
weather = { "temp" : [45,23,54,13,34,32,46],
           "humi" : [90,45,56,23,76,43,43],
            "wind" : [34,98,32,87,55,45,23],
            "Day" : ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "friday","saturday"]}
data = pd.DataFrame(weather)
print(data)

Highest_temp = np.max(data["temp"])
lowest_temp = np.min(data["temp"])
avg_temp = np.mean(data["temp"])
print(" Highest temperature :- ",Highest_temp)
print(" Lowest temperature :- ",lowest_temp)
print(" avg temperature :- ",avg_temp)

def temperature(Temp):
    if Temp < 20:
        return "cold"
    elif Temp <= 30:
        return "moderate"
    else:
        return "hot"

for i in range(len(data)):
    day = data.loc[i, "Day"]
    temp1 = data.loc[i, "temp"]
    category = temperature(temp1)
    print(day, ":", temp1, "°C -", category)    

```
